## Setup libraries

```python
from fastai.vision.all import *
from timm import create_model
```

## Get data

```python
#path = Path('src/png_nohist')
path = Path('/media/veracrypt2/Data Science/SIIM-Imgs/png_noFrameYesHist')
stage = 'train'

df = pd.read_pickle('src/train_ready_1024.pkl')
```

```python
DEBUG = False
```

```python
df.head(1)
```

Shrink the dataframe for prototyping

```python
if DEBUG:
    df = df.sample(n = 1500).reset_index(drop = True)
```

Get the paths of the image files

```python
items = get_image_files(path/'train')
items = L([fn for fn in items if fn.stem in df.id.tolist()])
```

```python
print(len(items))
```

Get functions for the dataloader

```python
def get_items(noop):
    return items

def get_label(fn):
    return df.loc[df['id'] == fn.stem, 'label'].values[0]
```

## Create DataLoaders

```python
size = 300

db = DataBlock(
    blocks = (ImageBlock(), CategoryBlock),
    get_items = get_items,
    splitter = RandomSplitter(),
    get_y = get_label,
    item_tfms = Resize((size, int(size * 1.16)), method = ResizeMethod.Squish),
    batch_tfms = [Normalize.from_stats(*imagenet_stats)]
)
```

```python
dls = db.dataloaders(path/stage, bs = 4, num_workers = 0)
```

```python
dls. show_batch()
```

## Create learner

```python
learn = cnn_learner(
    dls, 
    xresnet50,
    metrics = accuracy
)
```

```python
#learn.model
```

## Load pretrained weights
These weights come from training on a object detection problem, rather than clasification.

```python
load_model('xres50_bbox_enc.pkl', learn.model[0], opt=None, with_opt=False)
```

## Frozen / unfrozen training at given size

```python
learn.freeze()
```

```python
learn.lr_find()
```

```python
#newerdata_size: 300
learn.fit_one_cycle(5, lr_max = 1e-3)
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
#newerdata_size: 300
learn.fit_one_cycle(5, lr_max = slice(1e-6, 1e-4))
```

## Save the result
The encoder can be used in training for object detcetion. The learner is used to predict the test images.

```python
encoder = get_model(learn.model)[0]
save_model('data/xres50_enc.pkl', encoder, opt = None, with_opt=False)
learn.export('data/classifier_xres50.pkl')
```
